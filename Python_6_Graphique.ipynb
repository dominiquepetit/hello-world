```python
# La température d’ébullition de l’eau varie en fonction de la pression athmospéhérique
# Au niveau de la mer, la pression est en moyenne de 1013 hecto Pascal, noté hPa
# En haut du mont Blanc l'altitude est de 4 806 m et la pression seulement de 440 hPa, l’eau bout à 85°C
# En haut de l’Everest (altitude 8 849 m), la pression tombe à 330 hPa et l'eau bout à près de 72°C

# Importer le module graphique (plot)
import matplotlib.pyplot

# Les points à dessiner, sous forme de liste, une pour x, une pour y
x_pression_athmospherique_hPa = [1013, 800, 600, 400]     # abscisse
y_point_ebullition_eau_celcius = [100, 93.6, 86.0, 77.4]  # ordonnée

# Calcul et affichage du graphique
matplotlib.pyplot.plot(x_pression_athmospherique_hPa, y_point_ebullition_eau_celcius)   # plot() relie les points par défaut
```

```python
# Améliorons le graphique

# Importer le module avec un alias (as) pour simplifier l'écriture du code
import matplotlib.pyplot as plt

# Les points à dessiner, sous forme de liste, une pour x, une pour y
x_pression_athmospherique_hPa = [1013, 800, 600, 400]     # abscisse
y_point_ebullition_eau_celcius = [100, 93.6, 86.0, 77.4]  # ordonnée

# Ajoute des éléments de présentation
plt.title("Point d'ébullition de l'eau en fonction de la pression athmosphérique")
plt.xlabel("Pression de l'air (hPa)")
plt.ylabel("Point d'ébullition (°C)")

# Ajoute une grille
plt.grid()

# Calcul des éléments
# plot(liste_x, liste_y, options1, option2 ...)
plt.plot(x_pression_athmospherique_hPa, y_point_ebullition_eau_celcius, linestyle = "dotted")   # linestyle = 'dotted' : style de ligne = points
plt.plot(x_pression_athmospherique_hPa, y_point_ebullition_eau_celcius, "o", color="red")       # "o" : ajoute les points des valeurs

# Affichage du graphique
plt.show()
```

```python
# Equation du premier degré
# ax + b = 0

# Importer le module graphique
import matplotlib.pyplot as plt

# Importer le module mathématique
import numpy as np

# Definir les coefficients
a = 3
b = 5

# Générer les valeurs de x
# La fonction espacement linéaire du module numpy génère une liste de valeurs linspace(valeur_mini, valeur_maxi, nombre_de_valeurs)
# Prenons une liste de 100 valeurs réparties régulièrement entre -5 et 5
x = np.linspace(-5, 5, 100)

# Calculer les valeurs de y en fonction de x selon l'équation du second degré
y = a * x + b

# Ajoute les éléments de présentation
plt.title("ax + b = 0")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

# Créer la courbe
plt.plot(x, y)

# Afficher le graphique
plt.show()
```

```python
# Equation du second degré
# ax^2 + by + c = 0
# donc y = -(ax^2 + c) / b

# Importer le module graphique
import matplotlib.pyplot as plt

# Importer le module mathématique
import numpy as np

# Definir les coefficients
a = 3
b = 2
c = -3

# Générer les valeurs de x
# La fonction espacement linéaire du module numpy génère une liste de valeurs linspace(valeur_mini, valeur_maxi, nombre_de_valeurs)
# Prenons une liste de 100 valeurs réparties régulièrement entre -5 et 5
x = np.linspace(-5, 5, 100)

# Calculer les valeurs de y en fonction de x selon l'équation du second degré
y = -(a * x**2 + c) / b

# Ajoute des éléments de présentation du graphique
#plt.title("ax^2 + by + c = 0")
titre = str(a) + "x^2 + " + str(b) + "y + " + str(c) + " = 0"
plt.title(titre)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

# On peut forcer les valeurs min et max du graphique: axis([xmin, xmax, ymin, ymax])
plt.axis([-10, 10, -10, 10])

# Créer l'axe horizontal en pointillé
axe_x = [-10, 10]
axe_y = [0, 0]
plt.plot(axe_x, axe_y, linestyle = "dotted", color="black")

# Créer l'axe vertical
axe_x = [0, 0]
axe_y = [-10, 10]
plt.plot(axe_x, axe_y, linestyle = "dotted", color="black")

# Créer la courbe
plt.plot(x, y)

# Ajouter un point à mettre en valeur
x_point = 2.0
y_point = -(a * x_point**2 + c) / b
plt.scatter(x_point, y_point, color="red")

# Ajouter un commentaire pour le point
commentaire = "y = " + str(y_point)
plt.text(x_point + 0.5, y_point, commentaire)

# Ajouter une annotation avec une flèche
plt.annotate("Maximum", xy=(0, 1.7), xytext=(2,4), arrowprops=dict(facecolor="black", arrowstyle="->"))

# Afficher le graphique
plt.show()
```

# Deux séries de données

```python
#cosinus

import matplotlib.pyplot as plt
import numpy as np

# Générer 30 valeurs de x entre 0 et 2PI
x = np.linspace(0, 2*np.pi, 30)

# Générer les valeurs de y pour chaque valeur de x
y1 = np.cos(x)
y2 = np.sin(x)

# Créer les courbes
plt.plot(x, y1, label="cos(x)", linewidth=4)      # linewidth: largeur de la ligne
plt.plot(x, y2, label="sin(x)", color='purple')
plt.legend()                                      # Créer la légende pour afficher les labels

# Afficher le graphique
plt.show()
```

# Vecteurs

```python
import matplotlib.pyplot as plt
import numpy as np

# Premier vecteur
v1 = [1, 2]
origin_v1 = [0, 0]

# Deuxième vecteur
v2 = [4, 5]
origin_v2 = [1, 1]

# Créer les vecteurs avec la fonction quiver()
plt.quiver(*origin_v1, v1[0], v1[1], angles="xy", scale_units="xy", scale=1, color="blue", label="v")
plt.quiver(*origin_v2, v2[0], v2[1], angles="xy", scale_units="xy", scale=1, color="red", label="v2")

# Ajoute les éléments de présentation
plt.xlim(-1, 7)
plt.ylim(-1, 7)
plt.title("Vecteurs")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.legend()

# Affichage du graphique
plt.show()
```

# Histogramme : répartition d'une caractéristique

```python
import matplotlib.pyplot as plt

# Taille des plants en cm
plant_heights = {
    'Groupe A': [11, 52, 48, 9, 49, 50, 22, 54, 53, 47],
    'Groupe B': [42, 41, 39, 40, 37, 42, 36, 43, 35, 44],
    'Groupe C': [62, 65, 63, 64, 61, 62, 60, 67, 59, 63]
}

# Créer l'histogramme pour chaque groupe
for species, heights in plant_heights.items():
    plt.hist(heights, bins=5, edgecolor='white', label=species)

# Présentation
plt.xlabel('Hauteur (cm)')
plt.ylabel('Fréquence')
plt.title('Distribution de la hauteur des plants à maturité')
plt.grid()
plt.legend()

# Afficher
plt.show()
```

```python
import matplotlib.pyplot as plt

# Données de la population
age_groupes = ["0-9", "10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-79", "80+"]
homme_population = [5000, 4800, 4500, 4200, 4000, 3800, 3500, 3000, 2000]
femme_population = [4800, 4600, 4300, 4100, 3900, 3700, 3400, 2900, 1800]

# Fixer la taille de la figure
plt.figure(figsize=(10, 6))

# Créer les barres
plt.barh(age_groupes, homme_population, color="skyblue", label="Homme")
plt.barh(age_groupes, [-p for p in femme_population], color="pink", label="Femme")  # Valeurs négatives pour les bars vers la gauche

# Présentation
plt.xlabel("Population")
plt.ylabel("Groupe d'âge")
plt.title("Pyramide des âges")
plt.legend()
plt.tight_layout()

plt.show()
```

# Camembert (pie) : répartition en pourcentage

```python
import matplotlib.pyplot as plt

x = [37, 2, 30, 16, 15]
etiquette = ["Nucléaire", "Charbon", "Pétrole", "Gaz naturel", "Renouvelable et déchets"]
colors = ["purple", "black", "darkblue", "lightblue", "green"]
exploser = [0.1, 0, 0, 0, 0.1]

plt.title("Répartition énergétique de la France en 2023")
plt.pie(x, labels = etiquette, explode = exploser, colors=colors, shadow = True)

plt.show()
```

# Carte thermique (heatmap) : 3 données en 2D

```python
import matplotlib.pyplot as plt

xmin = -3
xmax = 3
nbx = 51
ymin = -2
ymax = 2
nby = 41

x = np.linspace(xmin, xmax, nbx)
y = np.linspace(ymin, ymax, nby)
X, Y = np.meshgrid(x, y)

# calcul du tableau des valeurs de Z
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)

# Préparer le graphique
fig, ax = plt.subplots()

# Créer la heatmap
im = ax.imshow(Z, interpolation="bicubic", origin="lower", extent=[xmin,xmax,ymin,ymax])

# Créer la graduation de la valeur Z
fig.colorbar(im)

plt.title("Objet 156-YU\nRayonnement X")  # \n pour le retour à la ligne

plt.show()
```

# Graphique 3D

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

plt.style.use('_mpl-gallery')

# Générer les données
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Créer la surface
fig, ax = plt.subplots(figsize=(5, 5), subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
